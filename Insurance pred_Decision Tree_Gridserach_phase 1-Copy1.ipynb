```python
import pandas as pd
```

```python
data=pd.read_csv("insurance_pre.csv")
```

```python
data
```

```python
data=pd.get_dummies(data,drop_first=True)
data
```

```python
data.columns
```

```python
indep=data[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dep=data[['charges']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
param={'criterion':['mse','mae','friedman_mse'],'max_features': ['auto','sqrt','log2'],'splitter':['best','random']}
grid=GridSearchCV(DecisionTreeRegressor(),param,refit=True,verbose=3,n_jobs=-1)
grid.fit(indep,dep)
```

```python
result=grid.cv_results_
print("R2,Best result combination{}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(result)
table
```

```python
age=float(input("Age:"))
BMI=float(input("BMI:"))
children=int(input("Children:"))
sex=int(input("Male-1 or Female-0:"))
smoker=int(input("Yes-0 or No-1:"))
```

```python
Future_charge_prediction=grid.predict([[age,BMI,children,sex,smoker]])
Future_charge_prediction
print("Future Prediction:{}".format(Future_charge_prediction))
```
