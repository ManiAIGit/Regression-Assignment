```python
import pandas as pd
```

```python
data=pd.read_csv("50_Startups.csv")
data
```

```python
data=pd.get_dummies(data,drop_first=True)
data
```

```python
data.columns
```

```python
indep=data[['R&D Spend', 'Administration', 'Marketing Spend',
       'State_Florida', 'State_New York']]
indep
```

```python
dep=data[['Profit']]
dep
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(indep,dep,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scx=StandardScaler()
x_train=scx.fit_transform(x_train)
x_test=scx.transform(x_test)
```

```python
from sklearn.preprocessing import StandardScaler
scy=StandardScaler()
y_train=scy.fit_transform(y_train)
y_test=scy.transform(y_test)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
param={'kernel':['linear','rbf','poly'],'C':[10,100,1000,2000,3000],'gamma':['auto','scale']}
grid=GridSearchCV(SVR(), param, refit=True, verbose=3, n_jobs=-1) 
grid.fit(x_train,y_train)
```

```python
result=grid.cv_results_
print("Best results{}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(result)
table
```

```python
y_pred=grid.predict(x_test)
from sklearn.metrics import r2_score
rscore=r2_score(y_test,y_pred)
rscore
```
